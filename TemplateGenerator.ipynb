```python
templates_params = mr.get_default_templates_params()
cell_models_folder = mr.get_default_cell_models_folder()
```

```python
amps = []
for t in tempgen.templates:
    amps.append(np.abs(np.min(t)))
amps = np.array(amps)
```

```python
_ = plt.hist(amps, bins=100)
```
